# Numerical Methods for root finding

## 1. Without visualization

```python
def f(x):
    return x**3-x-1
```

```python
def regula_falsi(f,a,b,tol,max_iter):
    if f(a)*f(b)>0:
        print("error taking a and b, their functional value must be opposite signs")
    iter=0
    while iter<max_iter:
        c= (a*f(b)-b*f(a))/(f(b)-f(a))
        if abs(f(c))<tol:
            return c
        elif f(a)*f(c)<0:
            a=c
        else:
            b=c
        print(f"iteration:{iter} x={c}")
        iter+=1
    return c
```

```python
regula_falsi(f,1,2,1e-4,100)
```

```python
def secant(f,a,b,tol,max_iter):
    iter=0
    while iter<max_iter:
        c= (a*f(b)-b*f(a))/(f(b)-f(a))
        if abs(f(c))<tol:
            return c
        else:
            a=b
            b=c
        print(f"iteration:{iter} x={c}")
        iter+=1
    return c
```

```python
secant(f,1,2,1e-4,100)
```

```python
def bisection_method(f, a, b, tol, max_iter):
    if f(a) * f(b) >= 0:
        raise ValueError("Function values at a and b must have opposite signs")
    
    c = a  # Initialize c before the loop
    iter = 0
    while (b - a) / 2 > tol and iter < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c  # Found exact solution
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
        print(f"iteration:{iter} x={c}")
        iter += 1
    
    return c  # c will always have a value here
        
```

```python
bisection_method(f,1,2,1e-4,100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - x - 1

# Define the range of x values
x = np.linspace(-2, 2, 1000)

# Calculate y values
y = f(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x³ - x - 1')
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Visualization of f(x) = x³ - x - 1')
plt.legend()
plt.grid(True)
plt.show()
```
